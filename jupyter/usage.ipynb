```python
import config
from navernews_10days import get_bow
from ekmeans import EKMeans, proportion_keywords

X, idx_to_vocab, vocab_to_idx = get_bow(date='2016-10-20', tokenize='noun')
X.shape
```

```python
n_clusters = 300
epsilon = 0.5
min_size = 5
init = 'random'
max_iter = 20
tol = 0.005
random_state = None
metric = 'cosine'
postprocessing=True
verbose = True
max_depth = 10

ekmeans = EKMeans(n_clusters, epsilon, max_depth, min_size,
        max_iter, tol, init, metric, random_state, postprocessing, verbose)

labels = ekmeans.fit_predict(X)
```

```python
import numpy as np

centers = ekmeans.cluster_centers_
labels = ekmeans.labels_
labels = labels[np.where(labels >= 0)[0]]
cluster_size = np.bincount(labels, minlength=centers.shape[0])

keywords = proportion_keywords(centers, labels, topk=20, candidates_topk=50, index2word=idx_to_vocab, min_score=0.3)
```

```python
for i, keywords_i in enumerate(keywords):
    if i % 5 == 4:
        print()
    keywords_strf = ', '.join(k for k, _ in keywords_i)
    size = cluster_size[i]
    print('[cluster {}, {} docs]: {}'.format(i, size, keywords_strf))
```
